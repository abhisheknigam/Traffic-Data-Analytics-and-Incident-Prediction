```python
import time
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
import numpy as np
from sklearn.svm import SVR

reg_iter = 1
total_error = []
total_error_svr = []
def linear_regression(x_train, y_train, x_test, y_test, x_predict):
    if len(x_train) < 2 or len(x_test) == 0 or (len(x_train) != len(y_train)) or (len(x_test) != len(y_test)) :
        return np.zeros(2)

    global reg_iter
    reg_iter = reg_iter + 1
#     if reg_iter == 4:
#         print(x_test)
#         print(y_test)
    #print('\ntraining regression for cols : ' + str(reg_iter))
    svr_model = SVR(C=1.0, epsilon=0.2)
    svr_model.fit(x_train, y_train)
    predicted_svr = svr_model.predict(x_test)

    regression_model = LinearRegression()
    regression_model.fit(x_train, y_train)
    #print('Coefficients: \n', regression_model.coef_)
    predicted_ = regression_model.predict(x_test)
#     for val in zip(predicted_, y_test):
#         print(str(val[0]) + '\t' + str(val[1]))
    error = np.mean((predicted_ - y_test) ** 2)
    total_error.append(error)
    
    error_svr = np.mean((predicted_svr - y_test) ** 2)
    total_error_svr.append(error_svr)
    
#     return regression_model.predict(x_predict)
    return svr_model.predict(x_predict)
```

```python
from datetime import date, datetime, timedelta

def perdelta(start, end, delta):
    curr = start
    while curr < end:
        yield curr
        curr += delta

```

```python
import time
import math
from numpy import array

total_error = []
lat_Long_Dict = {}
count = 0
start = time.time()
output_arr = []
def create_Lat_Long_Dict(row):
    columns1 = ['month1', 'year1', 'day1_actual', 'hour1']
    columns2 = ['month2', 'year2', 'day2_actual', 'hour2']
    
    df1 = pd.DataFrame(row[columns1])
    df1 = df1.transpose()
    df2 = pd.DataFrame(row[columns2])
    df2 = df2.transpose()
    df2.columns = columns1
    
    df1_arr = np.asarray(df1)
    df2_arr = np.asarray(df2)
    
    df_predict_arr = []
    
    for result in perdelta(datetime(df1_arr[0][1], df1_arr[0][0], df1_arr[0][2], df1_arr[0][3]), datetime(df2_arr[0][1], df2_arr[0][0], df2_arr[0][2], df2_arr[0][3]), timedelta(hours=1)):
        df_predict_arr.append(result)
   
    x_predict = pd.DataFrame(df_predict_arr,columns = ['start'])
    x_predict["start"] = pd.to_datetime(x_predict["start"])
    x_predict['month'] = x_predict['start'].dt.month
    x_predict['year'] = x_predict['start'].dt.year
    x_predict['day'] = x_predict['start'].dt.dayofweek
    x_predict['hour'] = x_predict['start'].dt.hour
    del x_predict['start']
    

    intermediate_output = []
    m =  df_event[(row["se_lat"] < df_event["latitude"]) & (row["nw_lat"] > df_event["latitude"]) & (row["se_lon"] > df_event["longitude"]) & (row["nw_lon"] < df_event["longitude"])]
    if(len(m) > 0):
        data = []

        data.append(m[m["event_type"] == "accidentsAndIncidents"])
        data.append(m[m["event_type"] == "roadwork"])
        data.append(m[m["event_type"] == "precipitation"])
        data.append(m[m["event_type"] == "deviceStatus"])
        data.append(m[m["event_type"] == "obstruction"])
        data.append(m[m["event_type"] == "trafficConditions"])
        
        
        for i in range(0,6,1):
            
            meta = data[i]
            if (len(meta) != 0):
                meta = meta.sort_values('for_prediction_timestamp')

                mHead = meta.head(1)
                mTail = meta.tail(1)

                m_df_arr = []
                for result in perdelta(datetime(mHead["for_prediction_timestamp"].dt.year, mHead["for_prediction_timestamp"].dt.month, mHead["for_prediction_timestamp"].dt.day, mHead["for_prediction_timestamp"].dt.hour), datetime(mTail["for_prediction_timestamp"].dt.year, mTail["for_prediction_timestamp"].dt.month, mTail["for_prediction_timestamp"].dt.day, mTail["for_prediction_timestamp"].dt.hour), timedelta(hours=1)):
                    m_df_arr.append(result)

                x_train_raw = pd.DataFrame(m_df_arr,columns = ['start'])
                x_train_raw["start"] = pd.to_datetime(x_train_raw["start"])
                
                x_train_raw['month'] = x_train_raw['start'].dt.month
                x_train_raw['year'] = x_train_raw['start'].dt.year
                x_train_raw['day'] = x_train_raw['start'].dt.dayofweek
                x_train_raw['day_actual'] = x_train_raw['start'].dt.day
                x_train_raw['hour'] = x_train_raw['start'].dt.hour
                del x_train_raw['start']
                cols = ['year','month','day','day_actual','hour']
                day_of_week_cols = ['month','year','day', 'day_actual','hour']

                all_cols = ['month','year','day','day_actual','hour','count']
                x_train_raw = x_train_raw.append(m[cols])


                x_train_raw = x_train_raw.drop_duplicates(subset=cols, keep=False)

                x_train_raw = x_train_raw[cols]
                x_train_raw = x_train_raw.append(m[all_cols])
                cur_data = x_train_raw;

                cur_data = cur_data.groupby(["year","month",'day','hour'])["count"].count().reset_index()
                
 
                columns = ['month', 'year','day', 'hour']
                
                data_test = cur_data[cur_data['year'] == 2014]
                data_train = cur_data[cur_data['year'] != 2014]
                x_train, y_train, x_test, y_test = data_train[columns] ,data_train['count'], data_test[columns], data_test['count']
                
                y = linear_regression(x_train, y_train, x_test, y_test, x_predict)
                intermediate_output.append(np.sum(y,axis=0))
            else:
                intermediate_output.append(0);
            
    output_arr.append(intermediate_output)   
    global count
    global start
    count = count + 1
    if count%2 == 0:
        end = time.time()
        print (str(count) + '\t' + str(end - start))
        start = time.time()

test = df_prediction_trial.head(10).apply(lambda row:create_Lat_Long_Dict(row), axis= 1)
output_arr_final = array(output_arr) 
df_final_output = pd.DataFrame(output_arr_final)
df_final_output.to_csv("SVR_year_hour.tsv",sep="\t")
print(math.sqrt(np.mean(total_error,axis=0)))
```
